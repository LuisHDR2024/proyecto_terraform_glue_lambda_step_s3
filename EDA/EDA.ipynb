```python
np.set_printoptions(threshold=np.inf)  # muestra todo

# explorando valores unicos de columnas categoricas

categorical_cols = \
["Order ID","Order Date","Ship Mode","Customer ID","Customer Name",
"Segment","Country","City","State","Region","Product ID",
"Category","Sub-Category","Product Name"]

for col in categorical_cols:
    print(f"--- {col} ---")
    print("Nº de únicos:", mydf[col].nunique())
    print("Primeros 10 valores únicos:", mydf[col].unique()[:20])
    
    # Top 10 más frecuentes
    print("Top 5 más frecuentes:")
    print(mydf[col].value_counts().head(5))
    
    # 10 menos frecuentes
    print("5 menos frecuentes:")
    print(mydf[col].value_counts().tail(5))
    
    print("\n")
```
